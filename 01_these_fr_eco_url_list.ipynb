```python
from lxml import html
import requests
import time
import pickle
```

```python
# Economie / auteurs
BASE_URL = 'https://www.theses.fr/fr/personnes/?q=&fq=dateSoutenance:[2000-01-01T23:59:59Z%2BTO%2B2022-12-31T23:59:59Z]&checkedfacets=role=auteur;oaiSetSpec=ddc:330;&start='
```

```python
pages = range(0, 17027, 10) # Incr√©mente par 10
```

```python
output = []
for page in pages:
    time.sleep(2)
    response = requests.get(BASE_URL + str(page))
    xml = html.document_fromstring(response.content)
    liens = xml.xpath('//div[@class = "personne_nom"]//a/@href')
    for lien in liens:
        output.append(lien)
    print(page)
    print(len(output))
```

```python
with open('pages_list.pickle', 'wb') as f:
    pickle.dump(output, f)
```
